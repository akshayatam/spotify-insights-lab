```python
from matplotlib import pyplot as plt 
import seaborn as sns 

%matplotlib inline 
```

```python
sns.set(style="whitegrid", palette="Greens_d") 
plt.rcParams["figure.figsize"] = (10, 6) 
```

```python
# Histogram of track durations in minutes
plt.hist(df["duration_min"], bins=50, color="green", edgecolor="black")
plt.title("Track Duration Distribution (in Minutes)")
plt.xlabel("Duration (min)")
plt.ylabel("Number of Tracks")
plt.grid(True)
plt.show()
```

```python
# Distribution plots for danceability, energy, and valence
features = ["danceability", "energy", "valence"]
for feature in features:
    sns.histplot(df[feature], kde=True, color="green")
    plt.title(f"{feature.capitalize()} Distribution")
    plt.xlabel(feature.capitalize())
    plt.ylabel("Density")
    plt.show()
```

```python
corr = df[[
    "danceability", "energy", "loudness", "speechiness", "acousticness",
    "instrumentalness", "liveness", "valence", "tempo", "duration_min"
]].corr()

sns.heatmap(corr, annot=True, cmap="Greens", fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap of Audio Features")
plt.show()
```

```python
# Explicit vs Non-Explicit track count by year
explicit_by_year = df.groupby(["year", "explicit"]).size().unstack(fill_value=0)
explicit_by_year.plot(kind="bar", stacked=True, color=["#a1d99b", "#31a354"]) 
plt.rcParams["figure.figsize"] = (20, 6) 
plt.title("Explicit vs Non-Explicit Tracks by Year")
plt.xlabel("Year")
plt.ylabel("Track Count")
plt.legend(["Non-Explicit", "Explicit"])
plt.tight_layout()
plt.show()
```

```python
# Scatter plot of Loudness vs Energy
sns.scatterplot(data=df, x="loudness", y="energy", alpha=0.5, color="green")
plt.title("Loudness vs Energy")
plt.xlabel("Loudness (dB)")
plt.ylabel("Energy")
plt.grid(True)
plt.show()
```

```python
import re
```

```python
# Function to fix release_date
def fix_release_date(val):
    if pd.isna(val):
        return None
    # Match exactly 4-digit year strings (e.g., "2001")
    if re.fullmatch(r"\\d{4}", str(val)):
        return f"{val}-01-01"
    return val

# Apply the fix
df["release_date"] = df["release_date"].apply(fix_release_date)

# Optional: Convert to datetime (ensures PostgreSQL-compatible format)
df["release_date"] = pd.to_datetime(df["release_date"], errors="coerce").dt.date
```

```python
# Define useful columns
columns_to_keep = [
    "id", "name", "album", "artists", "explicit",
    "danceability", "energy", "loudness", "speechiness",
    "acousticness", "instrumentalness", "liveness", "valence",
    "tempo", "duration_min", "year", "release_date"
]

# Filter columns
df_cleaned = df[columns_to_keep]

# Optional: Drop rows with missing name/album
df_cleaned = df_cleaned.dropna(subset=["name", "album"])

# Save to CSV for PostgreSQL ingestion
df_cleaned.to_csv("../data/spotify_tracks_cleaned.csv", index=False)
```
